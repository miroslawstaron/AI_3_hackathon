# Using the chat interface

Ollama framework provides several interfaces, one of them is the /api/generate, another one is the /api/chat. 

This workbook demonstrates how to use the /api/chat interface, both from the pure Python code and from the Gradio interface.

```python
def get_code(response):
    '''This method gets the program code generated by the prompt. 
       The generation often results in a markdown code block. 
       This method extracts the code from the markdown block and returns it as a string.'''
    # create a json object from the response
    resp = json.loads(response.text)

    # get the code from the response as a markdown
    mdCode = resp['message']['content']

    # remove the first and the last line
    mdCode = mdCode.split('\n')
    mdCode = mdCode[1:-1]
    mdCode = '\n'.join(mdCode)
    
    return mdCode
```

```python
import requests
import json

messages = [
            {"role": "system",
             "content": "You are a software that writes C programs based on prompts. Provides only the code, no description."},
            {"role": "user", 
             "content": "Generate a Fibonacci code"}
            ]

url = 'http://deepthought.cse.chalmers.se:80/api/chat'
data = {
    "model": "gemma2:2b",
    "messages": messages,
    "stream": False, 
    "temperature": 0.0,
    }
headers = {'Content-Type': 'application/json'}

response = requests.post(url, data=json.dumps(data), headers=headers)
```

```python
print(get_code(response))
```
